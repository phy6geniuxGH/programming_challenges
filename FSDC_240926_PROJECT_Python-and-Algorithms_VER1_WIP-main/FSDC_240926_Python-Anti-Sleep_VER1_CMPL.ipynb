```python
import pyautogui
import time
import numpy as np

upperBound = 120
lowerBound = 60

while True:
    stochastic1, stochastic2= np.random.randint(lowerBound,upperBound), np.random.randint(lowerBound,upperBound)
    print(stochastic1),print(stochastic2)
    if pyautogui.press('esc'):
        print("stop")
        break
    pyautogui.press('volumedown')
    time.sleep(stochastic1)
    pyautogui.press('volumeup')
    time.sleep(stochastic2)
    
```
