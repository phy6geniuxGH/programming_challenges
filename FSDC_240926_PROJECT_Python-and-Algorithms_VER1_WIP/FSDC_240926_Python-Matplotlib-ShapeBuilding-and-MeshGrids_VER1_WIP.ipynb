```python
################################################
# A Series of Matplotlib and Meshgrid Practices
################################################

############################
# Practice 1: Meshgrid Samples - 4 Plots
############################

import numpy as np
import matplotlib.pyplot as plt
import random
from mpl_toolkits.mplot3d import Axes3D

sx = 128
sy = 128

# Define the range for the mesh grid
x = np.linspace(-5, 5, sx)
y = np.linspace(-5, 5, sy)

# Create the mesh grid
X, Y = np.meshgrid(y, x)

# Define a function to plot on the mesh grid
Z = (np.cos(2*np.sqrt(X**2 + Y**2)))

# Coloring the second plot - Reshape Z, from (sx,sy) to (sx*sy, 1),add 1 to offset the negative values
z_array = Z.reshape(sx * sy,1)+1
# Normalize z_array, knowing that the values are from 0 - max(z_array)
z_array = (z_array/max(z_array))
# Initiate a sx*sy by 3 matrix (RGB for all data points), scatter plot's c = 0 (black)
zeros_array = np.zeros([sx*sy, 3])
# Assign the transposed z_array to the last column of zeros_array 
zeros_array[:,-1] = z_array.T

# Create a figure and set of subplots
figure, axs = plt.subplots(nrows = 2, ncols = 2, figsize=(5, 5))

# Plot the mesh grid 
axs[0,0].contourf(X, Y, Z, cmap='jet')
axs[0,0].set_aspect('equal')
axs[0,1].scatter(X, Y, c=zeros_array, alpha = z_array, s = 1)
axs[0,1].set_aspect('equal')
axs[1,1].plot(z_array, c = 'b', linewidth = 0.1)

# Add a 3D figure in the 3rd subplot (223 means 2 x 2 subplot, 3 slot from top-left plot)
ax3d = figure.add_subplot(223, projection='3d')
ax3d.plot_surface(X,Y,Z, cmap='viridis')
ax3d.set_title('3D view')
ax3d.set_aspect('equal')

plt.tight_layout()
plt.show()
```
