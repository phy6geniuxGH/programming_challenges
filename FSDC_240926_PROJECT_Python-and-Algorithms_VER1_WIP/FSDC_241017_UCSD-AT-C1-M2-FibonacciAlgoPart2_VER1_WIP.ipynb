## Problem Statement/s
Compute the n-th Fibonacci number modulo m.
- **Specifications**
    - **Input:** Integers n and m.
    - **Output:** n-th Fibonacci number modulo m.
- **Formatting**
    - **Input:** Integer n and m.
    - **Output:** $F_n$ mod $m$
    - **Constraints:** $1 \leq n \leq 10^{14}$, $2 \leq m \leq 10^3$
### Computer for:
$$F_n \, mod \, m$$

```python
# Huge Fibonacci Number modulo m - Using Memoization
import numpy as np
import time
import sys

# Increase integer string conversion
sys.set_int_max_str_digits(100000000)
# Increase the recursion limit
sys.setrecursionlimit(50000) 

# Dashboard
num = 115
modulo_num = 1000

# Memoization
def FibonacciMod(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = FibonacciMod(n-1, memo) + FibonacciMod(n-2, memo)
    return memo[n] % modulo_num
    
# Time your algorithm
start_time = time.time()

# Place the code here to be timed:
print(f' The {num}th Fibonacci number modulo {modulo_num} is: {FibonacciMod(num)}')

end_time = time.time()
diff = end_time - start_time
print(f' Code execution time is: {diff*1000} ms')

# Create a for loop to generate multiple tests:
start_time = time.time()

# Place the code here to be timed:

print(f' Multiple Tests: ')
np.random.seed(5)
fibo_num_mod_array = np.random.randint(2,10, 5).astype(float)
print(fibo_num_mod_array)
for i in range(len(fibo_num_mod_array)):
    print(f' The {fibo_num_mod_array[i]}th Fibonacci number modulo {modulo_num} is: {FibonacciMod(fibo_num_mod_array[i])}')

end_time = time.time()
diff = end_time - start_time
print(f' Code execution time is: {diff*1000} ms')
```

```python
# Huge Fibonacci Number modulo m - Sympy Library - Fastest
import numpy as np
import time
import sys
from sympy import fibonacci

# Increase integer string conversion
sys.set_int_max_str_digits(100000000)

# Dashboard
num = 28161241
modulo_num = 239

# Sympy Fibonacci Function
def Fibonacci_sympy_mod(n):
    return np.mod(fibonacci(num), modulo_num)
    
# Time your algorithm
start_time = time.time()

# Place the code here to be timed:
print(f' The {num}th Fibonacci number modulo {modulo_num} is: {Fibonacci_sympy_mod(num)}')

end_time = time.time()
diff = end_time - start_time
print(f' Code execution time is: {diff*1000} ms')

# Create a for loop to generate multiple tests:
start_time = time.time()

# Place the code here to be timed:

print(f' Multiple Tests: ')
np.random.seed(5)
fibo_num_mod_array = np.random.randint(2,10, 5).astype(float)
print(fibo_num_mod_array)
for i in range(len(fibo_num_mod_array)):
    print(f' The {fibo_num_mod_array[i]}th Fibonacci number modulo {modulo_num} is: {Fibonacci_sympy_mod(fibo_num_mod_array[i])}')

end_time = time.time()
diff = end_time - start_time
print(f' Code execution time is: {diff*1000} ms')
```
